```python
import pandas as pd
data = {
    'Name': ['ALICE', 'Bob', 'Charlie', 'DAvid', 'eve'],
    'Address': ['JL. Anggrek 12', 'Jl. Mawar No. 5', 'Jl. Melati, No.10 ','Jl. Tulip 3A', 'JL. Sakura 2B']
}

df = pd.DataFrame(data)
df
```

```python
# changing case
df['Name'] = df.Name.str.upper()
df.Address = df.Address.str.title()
df
```

```python
# removing extra spaces
df['Address'] = df.Address.str.strip()
df
```

```python
# breaking the string
df[['Street Name', 'No.']] = df.Address.str.split(',', expand=True)
```

```python
# combining strings - complete info
df['Complete Info'] = df.Name + ' lives in ' + df.Address
df
```

```python
# replacing substrings
df.Address = df.Address.str.replace('JL.', 'Street')
df.Address = df.Address.str.replace('Jl.', 'Street')
df
```

```python
# checking the availability of substrings
df[df.Address.str.contains('Anggrek')]
```

```python
# extraction using regular expressions
df['House number'] = df.Address.str.extract(r'(\d+)')
df
```

```python
# changing date format
date_data = ['12-01-2023','Jan 15 2023', 'Jan 20 2023']
df_date = pd.DataFrame({'Date': date_data})

print("Before the conversion:")
display(df_date)

df_date['Date'] = pd.to_datetime(df_date['Date'],format='mixed')
print("After the conversion")
display(df_date)
```
